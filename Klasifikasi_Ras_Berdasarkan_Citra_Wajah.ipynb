```python
import os, zipfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.applications import MobileNetV2, ResNet50
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# --- 1. PERSIAPAN DATA ---
ZIP_PATH = '/content/UTKFace.zip'
EXTRACT_DIR = '/content/utk_data'

if not os.path.exists(EXTRACT_DIR):
    with zipfile.ZipFile(ZIP_PATH, 'r') as zip_ref:
        zip_ref.extractall(EXTRACT_DIR)

# Parsing & Filtering (Hanya 3 Ras: Caucasoid, Negroid, Mongoloid)
# Mapping: 0: Caucasoid, 1: Negroid, 2: Mongoloid
RACE_MAP = {0: 'Caucasoid', 1: 'Negroid', 2: 'Mongoloid'}
data = []
LIMIT_PER_RACE = 2000 # Agar seimbang dan training lebih cepat

img_folder = os.path.join(EXTRACT_DIR, 'UTKFace')
if not os.path.exists(img_folder): img_folder = EXTRACT_DIR

files = os.listdir(img_folder)
np.random.seed(42)
np.random.shuffle(files)

counts = {0: 0, 1: 0, 2: 0}
for fn in files:
    parts = fn.split('_')
    if len(parts) >= 3:
        try:
            r_id = int(parts[2])
            if r_id in RACE_MAP and counts[r_id] < LIMIT_PER_RACE:
                data.append({'filename': os.path.join(img_folder, fn), 'label': RACE_MAP[r_id]})
                counts[r_id] += 1
        except: continue

df = pd.DataFrame(data)
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['label'])

# --- 2. DATA GENERATOR ---
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2, horizontal_flip=True)
train_gen = datagen.flow_from_dataframe(train_df, x_col='filename', y_col='label', target_size=(128,128), batch_size=32, subset='training')
val_gen = datagen.flow_from_dataframe(train_df, x_col='filename', y_col='label', target_size=(128,128), batch_size=32, subset='validation')
test_gen = ImageDataGenerator(rescale=1./255).flow_from_dataframe(test_df, x_col='filename', y_col='label', target_size=(128,128), batch_size=32, shuffle=False)

# --- 3. IMPLEMENTASI 3 MODEL ---

# A. Model Base (Mengacu pada Arsitektur Jurnal Anda)
def build_base_cnn():
    model = Sequential([
        Conv2D(4, (3,3), activation='relu', padding='same', input_shape=(128,128,3)),
        Conv2D(8, (3,3), activation='relu', padding='same'),
        MaxPooling2D(2,2),
        Conv2D(16, (3,3), activation='relu', padding='same'),
        Conv2D(32, (3,3), activation='relu', padding='same'),
        MaxPooling2D(2,2),
        Flatten(),
        Dense(512, activation='relu'),
        Dense(3, activation='softmax') # Output 3 Kelas
    ])
    return model

# B. Model Pretrained 1 (MobileNetV2)
def build_mobilenet():
    base = MobileNetV2(weights='imagenet', include_top=False, input_shape=(128,128,3))
    base.trainable = False
    x = GlobalAveragePooling2D()(base.output)
    x = Dense(128, activation='relu')(x)
    out = Dense(3, activation='softmax')(x)
    return Model(inputs=base.input, outputs=out)

# C. Model Pretrained 2 (ResNet50)
def build_resnet():
    base = ResNet50(weights='imagenet', include_top=False, input_shape=(128,128,3))
    base.trainable = False
    x = GlobalAveragePooling2D()(base.output)
    x = Dense(128, activation='relu')(x)
    out = Dense(3, activation='softmax')(x)
    return Model(inputs=base.input, outputs=out)

# --- 4. PELATIHAN & EVALUASI ---
models = {"CNN_Base": build_base_cnn(), "MobileNetV2": build_mobilenet(), "ResNet50": build_resnet()}
summary_report = []

for name, model in models.items():
    print(f"\n" + "="*50)
    print(f" MEMULAI PROSES UNTUK MODEL: {name} ")
    print("="*50)

    # Compile Model
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    # Training
    history = model.fit(train_gen, validation_data=val_gen, epochs=10)

    # --- A. PLOT GRAFIK LOSS DAN ACCURACY ---
    plt.figure(figsize=(12, 4))

    # Grafik Akurasi
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy', marker='o')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='x')
    plt.title(f'Grafik Akurasi: {name}')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    # Grafik Loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss', marker='o')
    plt.plot(history.history['val_loss'], label='Validation Loss', marker='x')
    plt.title(f'Grafik Loss: {name}')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # --- B. PREDIKSI DATA TEST ---
    test_gen.reset()
    preds = model.predict(test_gen)
    y_pred = np.argmax(preds, axis=1)
    y_true = test_gen.classes
    class_names = list(test_gen.class_indices.keys())

    # --- C. CLASSIFICATION REPORT ---
    print(f"\n[HASIL] Classification Report - {name}:")
    # Mencetak Accuracy, Precision, Recall, dan F1-score untuk setiap kelas
    print(classification_report(y_true, y_pred, target_names=class_names))

    # --- D. CONFUSION MATRIX ---
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix: {name}')
    plt.ylabel('Label Aktual')
    plt.xlabel('Label Prediksi')
    plt.show()

    # Menyimpan ringkasan metrik untuk tabel akhir
    report_dict = classification_report(y_true, y_pred, output_dict=True)
    summary_report.append({
        'Model': name,
        'Accuracy': report_dict['accuracy'],
        'Precision': report_dict['weighted avg']['precision'],
        'Recall': report_dict['weighted avg']['recall'],
        'F1-Score': report_dict['weighted avg']['f1-score']
    })

    # Optional: Simpan model secara otomatis
    model.save(f"{name}.h5")
    print(f"\n[INFO] Model {name} telah disimpan sebagai {name}.h5")

# --- 5. TABEL ANALISIS AKHIR ---
df_summary = pd.DataFrame(summary_report)
print("\n" + "!"*50)
print(" TABEL PERBANDINGAN AKHIR ".center(50, "!"))
print("!"*50)
print(df_summary.to_string(index=False))
```

```python
%%writefile app.py
import streamlit as st
import tensorflow as tf
import numpy as np
import pandas as pd
from PIL import Image
import plotly.graph_objects as go

# --- KONFIGURASI HALAMAN ---
st.set_page_config(
    page_title="Race Classification AI",
    page_icon="👤",
    layout="wide"
)

# --- STYLE CSS KUSTOM ---
st.markdown("""
    <style>
    .main {
        background-color: #f5f7f9;
    }
    .stButton>button {
        width: 100%;
        border-radius: 5px;
        height: 3em;
        background-color: #007bff;
        color: white;
    }
    .result-box {
        padding: 20px;
        border-radius: 10px;
        background-color: white;
        box-shadow: 0 4px 6px rgba(0,0,0,0.1);
    }
    </style>
    """, unsafe_allow_html=True)

# --- FUNGSI LOAD MODEL ---
@st.cache_resource
def load_all_models():
    # Ganti nama file sesuai dengan file .h5 Anda
    cnn = tf.keras.models.load_model('CNN_Base.h5')
    mnet = tf.keras.models.load_model('MobileNetV2.h5')
    res = tf.keras.models.load_model('ResNet50.h5')
    return cnn, mnet, res

# --- SIDEBAR ---
with st.sidebar:
    st.image("https://cdn-icons-png.flaticon.com/512/3829/3829733.png", width=100)
    st.title("Navigation")
    page = st.radio("Go to", ["Dashboard Prediksi", "Analisis Performa Model"])

    st.info("""
    **Identitas:**
    - Nama: Asya Cahya Pradana
    - NIM: 202210370311041
    """)

# --- HALAMAN 1: DASHBOARD PREDIKSI ---
if page == "Dashboard Prediksi":
    st.title("👤 Sistem Klasifikasi Ras Manusia")
    st.write("Unggah citra wajah untuk mendeteksi etnis berdasarkan model Deep Learning.")

    col1, col2 = st.columns([1, 1])

    with col1:
        st.subheader("📤 Input Citra")
        uploaded_file = st.file_uploader("Pilih gambar wajah...", type=["jpg", "jpeg", "png"])

        selected_model_name = st.selectbox(
            "Pilih Model Klasifikasi",
            ["CNN_Base", "MobileNetV2", "ResNet50"]
        )

    if uploaded_file is not None:
        image = Image.open(uploaded_file)
        with col1:
            st.image(image, caption='Original Image', use_container_width=True)

        with col2:
            st.subheader("🎯 Hasil Prediksi")
            if st.button("Jalankan Prediksi"):
                with st.spinner('Menganalisis fitur wajah...'):
                    # Load model yang dipilih
                    cnn, mnet, res = load_all_models()
                    model_dict = {"CNN_Base": cnn, "MobileNetV2": mnet, "ResNet50": res}
                    model = model_dict[selected_model_name]

                    # Preprocessing
                    img = image.resize((128, 128))
                    img_array = tf.keras.preprocessing.image.img_to_array(img)
                    img_array = np.expand_dims(img_array, axis=0)
                    img_array = img_array / 255.0

                    # Prediksi
                    predictions = model.predict(img_array)
                    labels = ['Caucasoid', 'Negroid', 'Mongoloid']
                    result_idx = np.argmax(predictions)
                    confidence = np.max(predictions) * 100

                    # Tampilan Output
                    st.markdown(f"""
                    <div class="result-box">
                        <h2 style='text-align: center; color: #007bff;'>{labels[result_idx]}</h2>
                        <p style='text-align: center;'>Confidence Level: <b>{confidence:.2f}%</b></p>
                    </div>
                    """, unsafe_allow_html=True)

                    # Bar Chart Probabilitas
                    fig = go.Figure([go.Bar(x=labels, y=predictions[0]*100, marker_color='#007bff')])
                    fig.update_layout(title="Probabilitas Per Kelas (%)", height=300)
                    st.plotly_chart(fig, use_container_width=True)

# --- HALAMAN 2: ANALISIS PERFORMA ---
else:
    st.title("📊 Analisis Performa Model")
    st.write("Perbandingan metrik evaluasi berdasarkan hasil pengujian (Accuracy, F1-Score).")

    # Data dari hasil training Anda
    data = {
        'Model': ['CNN_Base', 'MobileNetV2', 'ResNet50'],
        'Accuracy': [0.8675, 0.7666, 0.5683],
        'F1-Score': [0.8674, 0.7672, 0.5534]
    }
    df = pd.DataFrame(data)

    col_metrics1, col_metrics2, col_metrics3 = st.columns(3)
    col_metrics1.metric("Best Accuracy", "86.75%", "CNN_Base")
    col_metrics2.metric("Avg MobileNet", "76.66%")
    col_metrics3.metric("Avg ResNet", "56.83%", "-19.83%")

    st.table(df)

    # Visualisasi Bar Chart
    fig = go.Figure()
    fig.add_trace(go.Bar(x=df['Model'], y=df['Accuracy'], name='Accuracy', marker_color='#007bff'))
    fig.add_trace(go.Bar(x=df['Model'], y=df['F1-Score'], name='F1-Score', marker_color='#ff7f0e'))
    fig.update_layout(barmode='group', title="Visualisasi Akurasi vs F1-Score")
    st.plotly_chart(fig, use_container_width=True)

    st.warning("""
    **Kesimpulan Analisis:** CNN Base (Non-Pretrained) mengungguli model raksasa karena dilatih khusus pada domain wajah manusia dengan parameter yang selaras dengan resolusi input.
    """)
```
